### Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('../data/nasa_firms_data.json')  # or pd.read_csv if you switch to CSV

# Basic info
print(f"First 5 rows of the DataFrame: \n{df.head()}")
```

```python
print(f"DataFrame info: \n{df.info()}")
```

```python
print(f"DataFrame description: \n{df.describe()}")
```

```python
# Check for missing values
print(df.isnull().sum())

# Visualize distributions, locations, etc.
import matplotlib.pyplot as plt
df['latitude'].hist()
plt.show()
```

```python
print(f"Data types: \n{df.dtypes}")
```

```python
# Visualize Distributions
df['brightness'].hist()
plt.title('Brightness Distribution')
plt.xlabel('Brightness')
plt.ylabel('Frequency')
plt.show()
```

```python
# Explore Relationships
plt.scatter(df['brightness'], df['frp'])
plt.xlabel('Brightness')
plt.ylabel('FRP')
plt.title('Brightness vs FRP')
plt.show()
```

```python
# Geographical Visualization
plt.scatter(df['longitude'], df['latitude'], c=df['brightness'], cmap='hot', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Fire Locations (colored by brightness)')
plt.colorbar(label='Brightness')
plt.show()
```

```python
print(df.groupby('daynight')['brightness'].mean())
```
